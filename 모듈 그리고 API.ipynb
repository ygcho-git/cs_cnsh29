### 방정식 풀이 모듈 : sympy

````python
from sympy import Symbol, solve

x = Symbol('x')
equation = 2 * x - 6

r = solve(equation)

print(r)
````

````python
from sympy import Symbol, solve

equation = x ** 2 + 3 * x + 2

r = solve(equation, dict=True) #dict 타입으로

print(r)
````

````python
from sympy import Symbol, solve

equation = x ** 2 + x + 1

r = solve(equation, dict=True)

print(r)      # 허수는 I로...
````

````python
# 연립 방정식 풀이

from sympy import Symbol, solve

y = Symbol('y')
equation1 = x + 2 * y - 3
equation2 = 3 * x - y + 2
solve((equation1, equation2), dict=True)
````
